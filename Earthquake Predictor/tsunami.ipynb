```python
import pandas as pd
import numpy as np
import math
```

```python
tsunami = pd.read_csv('tsunami_dataset.csv')
#tsunami.tail()
res_set = {'Very Doubtful Tsunami': 0, 'Questionable Tsunami': 1,
           'Definite Tsunami': 3, 'Probable Tsunami': 2,
           'Event that only caused a seiche or disturbance in an inland river': -1}

result = [res_set[row['EVENT_VALIDITY']] for index, row in tsunami.iterrows()]
result
tsunami.insert(21,"EVENT_OCCURRED",result,True)
tsunami
```

```python
tsunami['EVENT_VALIDITY'].unique()
```

```python
from sklearn.model_selection import train_test_split
tsunami_1 = tsunami
tsunami_1['EQ_MAGNITUDE'] = tsunami_1['EQ_MAGNITUDE'].fillna(tsunami['EQ_MAGNITUDE'].mean())
tsunami_2 = tsunami
tsunami_2['EQ_MAGNITUDE'] = tsunami_2['EQ_MAGNITUDE'].fillna(tsunami['EQ_MAGNITUDE'].median())
tsunami_1
```

```python
eq_depth_median = tsunami['EQ_DEPTH'].median()
eq_depth_mean = tsunami['EQ_DEPTH'].mean()
eq_depth_mean
eq_depth_median
```

```python
tsunami_11= tsunami_12 = tsunami_21 = tsunami_22 = tsunami
tsunami_11['EQ_DEPTH']  = tsunami_1.EQ_DEPTH.fillna(eq_depth_mean)
tsunami_12['EQ_DEPTH']  = tsunami_1.EQ_DEPTH.fillna(eq_depth_median)
tsunami_21['EQ_DEPTH']  = tsunami_2.EQ_DEPTH.fillna(eq_depth_mean)
tsunami_22['EQ_DEPTH']  = tsunami_2.EQ_DEPTH.fillna(eq_depth_median)
X = tsunami_11[['LATITUDE','LONGITUDE','EQ_DEPTH','EQ_MAGNITUDE']]
y = tsunami_11['EVENT_OCCURRED']


```

```python
from sklearn.ensemble import RandomForestRegressor

X_train, X_val, y_train, y_val = train_test_split(X,y)
y_train
```

```python
model = RandomForestRegressor(random_state=0, n_estimators = 200)
model.fit(X_train,y_train)

preds = model.predict(X_val)
preds = [math.ceil(x) for x in preds]
```

```python
from sklearn.metrics import mean_absolute_error

vals = mean_absolute_error(y_val,model.predict(X_val))
vals
```

```python
frequency_3 = {}
frequency_2 = {}
frequency_1 = {}
frequency_0 = {}

# Iterate through each row in the DataFrame
for index, row in tsunami.iterrows():
    region = row['REGION']
    event_occurred = row['EVENT_OCCURRED']
    
    if event_occurred == 3:
        if region not in frequency_3:
            frequency_3[region] = 0
        frequency_3[region] += 1
    elif event_occurred == 2:
        if region not in frequency_2:
            frequency_2[region] = 0
        frequency_2[region] += 1
    elif event_occurred == 1:
        if region not in frequency_1:
            frequency_1[region] = 0
        frequency_1[region] += 1
    elif event_occurred == 0:
        if region not in frequency_0:
            frequency_0[region] = 0
        frequency_0[region] += 1

frequency_3
```

```python
avg = (frequency_3['Japan']*3+frequency_2['Japan']*2+frequency_1['Japan']*1)/(frequency_3['Japan']+frequency_2['Japan']+frequency_1['Japan'])
```
